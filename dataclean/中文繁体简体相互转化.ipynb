```python
inputFile = "/Users/sunlu/Workspaces/PyCharm/PythonDiary/data/t2c._test.txt"
outputFile = "/Users/sunlu/Workspaces/PyCharm/PythonDiary/data/output_test.txt"
with open(outputFile, 'w') as opt:
    inp = open(inputFile, 'r')
    for line in inp:
        newline = HanziConv.toSimplified(line)
        opt.write(newline.encode('utf-8'))
        print(newline)
    inp.close()
```

读取wiki.zh.txt文件中的前50行进行繁简体转化

```python
filePath = "/Users/sunlu/Workspaces/PyCharm/Github/wiki_zh_word2vec/wiki.zh.txt"
fpa=open(filePath)
txt = []
indexx=0
for linea in fpa.readlines():
    indexx=indexx+1
    txt = txt + [linea]
    if indexx ==50:
        break
fpa.close()
outputFile = "/Users/sunlu/Workspaces/PyCharm/PythonDiary/data/output_wiki.txt"
with open(outputFile, 'w') as opt:
    for line in txt:
        newline = HanziConv.toSimplified(line)
        opt.write(newline.encode('utf-8'))
```
