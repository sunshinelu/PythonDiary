# 对dataframe中的字符串进行过滤

```python
import pandas as pd
df = pd.DataFrame([{'col1':'abc', 'col2':None}, 
                    {'col1':'b', 'col2':2},
                   {'col1':'acd', 'col2':3}, 
                   {'col1':'ac', 'col2':5},])
df
```

```python
df1 = df[df["col1"].str.contains("a")]
df1
```

```python
df2 = df[df['col1'].str.match("a")]
df2
```

```python
df2 = df[~df['col1'].str.match("a")]
df2
```

```python
df["col1"].isin(['ac'])
```

```python
df[df["col1"].isin(['ac'])]
```

```python
def find_udf(x):
    i = find(x)
    if i >= 0:
        result = True 
    else:
        result = False
    return result
df2 = df[df["col1"].apply(find_udf("a"))]
df2
```
