```python
def mean_data(self):
    # 插补均值
    col_mean = self.iloc[1:-1]
    fill_mean = self.fillna(col_mean.mean())
    return fill_mean
```

```python
def median_data(self):
    # 插补中位数
    col_median = self.iloc[1:-1]
    fill_median = self.fillna(col_median.median())
    return fill_median
```

```python
fill_median = median_data(data)
fill_median
```

```python
def mode_data(self):
    # 插补众数
    col_mode = self.iloc[1:-1]
    fill_mode = self.fillna(col_mode.median())
    return fill_mode
```

```python
fill_mode = mode_data(data)
fill_mode
```
