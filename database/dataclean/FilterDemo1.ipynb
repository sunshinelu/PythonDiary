# 在python中创建Dataframe并对DataFrame中的列进行过滤

## 1. 创建DataFrame

```python
import pandas as pd
import numpy as np
```

```python
data={'col1':['a','b','c'],'col2':[1,2,3]}
df = pd.DataFrame(data)
df
```

过滤col1中含有"a"的列

```python
df1 = df[df.col1.str.contains("a")]
df1
```

```python
df2 = df[df.col1.str.contains('a | b')]
df2
```

```python
df2 = df[df.col1.str.contains("a") | df.col1.str.contains("b")]
df2
```

过滤col2列中，大于等于2的行

```python
df3 = df[df.col2 >= 2]
df3
```

```python
columns = ["col1"]
df4 = df[columns]
df4
```

```python
s = "【天猫双11交易额破800亿元】"
data={'col1':[s,'b','c'],'col2':[1,2,3]}
df1 = pd.DataFrame(data)
df1
```

```python
df2 = df1[df1.col1.str.contains('\【')]
df2
```
